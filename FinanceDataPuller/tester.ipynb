```python
import pandas as pd
import numpy as np
import yfinance as yf
import time
from pytrends.request import TrendReq
```

```python
years = np.arange(2015, 2023)
results = []
for i in years:
    i = int(i)
    df = pd.read_csv(f'FinanceDataPuller/pytrendsCSV/pytrends_{i}.csv')
    df = df.drop('isPartial')
    df['date'] = pd.to_datetime(df['date'])
    df.index = df['date']
    for col in df.columns:
        top3 = df[col].nlargest(3)
        for date, value in top3.items():
            results.append({
                "year": i,
                "keyword": col,
                "date": date,
                "value": value
            })

# Convert to DataFrame
result_df = pd.DataFrame(results)
result_df.head()

        

```
